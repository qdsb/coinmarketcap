```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from time import sleep
```

```python
data = pd.read_csv('CoinsAndLinks.csv')
data.head()
```

```python
driver = webdriver.Firefox()
driver.maximize_window()
Ranks = []
Names =[]
Symbols = []
Price = []
MarketCaps = []
Volume_24Hr = []
Circulating_Supply = []
driver.get('https://coinmarketcap.com/historical/20230825/')
for i in range(1 ,201):
    Ranks.append(i)
    driver.execute_script(f"window.scrollTo(0 , {i} * 47)")
    sleep(1)
    Name_Coin = driver.find_element(By.XPATH , f'//*[@id="__next"]/div[2]/div[2]/div/div[1]/div[3]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[2]/div/a[2]').text
    Names.append(Name_Coin)
    Symbol_Coin = driver.find_element(By.XPATH , f'//*[@id="__next"]/div[2]/div[2]/div/div[1]/div[3]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[3]/div').text
    Symbols.append(Symbol_Coin)
    Price_Coin = driver.find_element(By.XPATH , f'//*[@id="__next"]/div[2]/div[2]/div/div[1]/div[3]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[5]/div').text
    Price.append(float(Price_Coin.replace('$' , '').replace(',' , '')))
    MarketCap_Coin = driver.find_element(By.XPATH , f'//*[@id="__next"]/div[2]/div[2]/div/div[1]/div[3]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[4]/div').text
    MarketCaps.append(float(MarketCap_Coin.replace('$' , '').replace(',' , '')))   
    Volume_Coin = driver.find_element(By.XPATH , f'//*[@id="__next"]/div[2]/div[2]/div/div[1]/div[3]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[7]/a').text
    Volume_24Hr.append(float(Volume_Coin.replace('$' , '').replace(',' , '')))
    Circulating_Coin = driver.find_element(By.XPATH , f'//*[@id="__next"]/div[2]/div[2]/div/div[1]/div[3]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[6]/div').text
    Circulating_Supply.append(int(Circulating_Coin.split()[0].replace(',','')))
```

```python
driver = webdriver.Firefox()
driver.maximize_window()
GitHub_Links = []
Tags = []
for i in range(1 , 201):
    driver.get(data.loc[i-1 , 'MainLink'])    
    GitHub_Flag = False
    All_Links = driver.find_elements(By.TAG_NAME , 'a')
    for Link in All_Links:
        try:
            if Link.text == 'GitHub':
                GitHub_Links.append(Link.get_attribute('href'))
                GitHub_Flag = True
        except:
            continue
    if GitHub_Flag == False:
        GitHub_Links.append('No GitHub Link')
    temp = []
    try:
        Tag_Element = driver.find_element(By.XPATH , '//*[@id="__next"]/div[2]/div[1]/div[2]/div/div/div[2]/div[3]/section[2]/div/div[7]')
        Tags_Coin = Tag_Element.find_elements(By.TAG_NAME, 'a')
        for tag in Tags_Coin:
            temp.append(tag.text)    
        Tags.append(temp)
    except:
        try:
            Tag_Element = driver.find_element(By.XPATH , '//*[@id="__next"]/div[2]/div[1]/div[2]/div/div/div[2]/div[4]/section[2]/div/div[7]')
            Tags_Coin = Tag_Element.find_elements(By.TAG_NAME , 'a')
            for tag in Tags_Coin:
                temp.append(tag.text)
            Tags.append(temp)
        except:
            Tags.append('No Tags')
driver.close()    
```

```python
column_Names = ['Rank' , 'Name' , 'Symbol' , 'Price' , 'Market Cap' , 'Volume (24 Hour)' , 'Circulating Supply' , 'Github Link' , 'Tags']
Full_Data = pd.DataFrame(list(zip(Ranks , Names , Symbols , Price , MarketCaps , Volume_24Hr , Circulating_Supply , GitHub_Links , Tags)) , columns =column_Names)
Full_Data
```
